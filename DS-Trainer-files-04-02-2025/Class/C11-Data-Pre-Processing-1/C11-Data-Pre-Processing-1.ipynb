```python
# Add a duplicate row of an existing row item

pd.concat([df, df.iloc[[-1]]])
```

```python
df_duplicates_explain = pd.concat([df, df.iloc[[-1]]])
```

```python
# Before resetting index
df_duplicates_explain.tail()
```

```python
df_duplicates_explain = df_duplicates_explain.reset_index(drop=True)
```

```python
# After resetting index

df_duplicates_explain.tail()
```

```python
df_duplicates_explain["Age"].drop_duplicates(keep="last")
```

```python
df_duplicates_explain["Age"].drop_duplicates(keep="last").index
```

```python
# Within df.loc[] function, pass the above obtained row numbers.
# The result is, it drops the duplicate "ID" row that is NOT LAST

required_index_after_dropping_duplicates = df_duplicates_explain["Age"].drop_duplicates(keep="last").index

df_duplicates_explain.loc[required_index_after_dropping_duplicates]
```
