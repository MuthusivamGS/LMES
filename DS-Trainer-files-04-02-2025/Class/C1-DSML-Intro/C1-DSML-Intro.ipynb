### AI vs ML vs DS vs DL

**DS  Data Science**
-	**A field of study** involving concepts from Mathematics (Statistics, Linear Algebra, etc.) and Computing (Python, GPUs, etc.)
-	Gives Fundamental theoretical knowledge for all data handling for Analysis and Predictions:
    -   Data Analysis
    -	ML
    -	DL
    -	AI

**ML  Machine Learning**
-	A branch of Data Science to create AI systems
-	Uses Statistics and other Maths concepts to create Algorithms
-	Those Algorithms are used to create Models  called **ML models**
-	Used for **predictive tasks involving Numerical data**


**DL  Deep Learning**
-	A branch of Data Science to create AI systems
-	Uses **Advanced Machine Learning algorithms**, based on Artificial Neural Networks
-	Those Algorithms are used to create Models  called DL models
-	Used for **predictive tasks involving unstructured and complex datasets such as Text, Image, Audio, and Video**
    - TEXT DATA
        - NLP - Natural Language Processing
            - LLM - Large Language Models
            - Eg: Chat-GPT


**AI - Artificial Intelligence**
-	A marketing term referring to human-like intelligent machines  not to worry about too much!
-	Usage of ML and DL concepts to build AI systems
-	Thats it.
