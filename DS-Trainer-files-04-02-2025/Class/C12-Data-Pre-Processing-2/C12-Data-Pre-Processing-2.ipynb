### [45m] Data Stored in Memory - 2 Important concepts

1. **Mutable** (list, dict, set) Vs **Immutable** (str, tuple)
2. **Copy** Vs **DeepCopy**

#### 1. **Mutable** Vs **Immutable**

```python
# List is Mutable
# That means, the elements within that Data type can be changed

my_list = [1,2,3]
print(my_list)
print(id(my_list))

print('----------------')

my_list[1] = 100
print(my_list)
print(id(my_list))

print('----------------')

my_list[0] = 1234
print(my_list)
print(id(my_list))
```

```python
print(id(my_list))
```

```python
print(id(my_list[0]))
```

```python
# Set is Mutable
# That means, the elements within that Data type can be changed

my_set = {1,2,3}
print(my_set)

# my_set[1] = 100
# print(my_set)
# TypeError: 'set' object does not support item assignment

print(id(my_set)) # Memory id value of my_set
```

```python
my_set.remove(2) # Removing an element from my_set
print(my_set)
print(id(my_set)) # Memory id value of my_set is the same. Which means, the memory id value of the variable is retained.
```

```python
my_set.add(100) # Adding an element to my_set
print(my_set)
print(id(my_set)) # Memory id value of my_set is the same. Which means, the memory id value of the variable is retained.
```

```python
# Str is Immutable
# That means, the elements within that Data type CANNOT be changed

my_string = "abcd"
print(my_string)
```

```python
my_string[0] = "w"
print(my_string)

#TypeError: 'str' object does not support item assignment
```

```python
print(id(my_string))
```

#### 2. **Copy** Vs **DeepCopy**

```python
import copy
# copy.copy()
# copy.deepcopy()
```

```python
variable1 = ["abc", True, 123, [444, "xyz"]]  # These are list items assigned to variable1

variable2 = copy.copy(variable1)              # SHALLOW COPY of variable1
variable3 = copy.deepcopy(variable1)          # DEEP COPY of variable1

variable4 = variable1                         # Simple assignment of variable1. It has same memory id value as variable1.

print("variable1:", variable1, id(variable1))
print("variable2:", variable2, id(variable2))
print("variable3:", variable3, id(variable3))
print("variable4:", variable4, id(variable4))

# So far the variable1 elements have been copied successfully to variable2, variable3, and variable4.
```

```python
variable1[2] = 111     # In variable1, Making a change to a outer list item
variable1[3][0] = 999  # In variable1, Making a change to a DEEP inner list item

print("variable1:", variable1, id(variable1))
print("variable2:", variable2, id(variable2))      #<<<<<<------ The result of doing only SHALLOW COPY of variable1
print("variable3:", variable3, id(variable3))      #<<<<<<------ The result of doing DEEP COPY of variable1
print("variable4:", variable4, id(variable4))      #<<<<<<------ The result of simple assignment of variable1

# The Results can be seen as:
# Due to above changes made to variable1,
# For variable2 - because SHALLOW COPY of variable1 was done earlier, the DEEP inner list item GOT CHANGED.
# For variable3 - because DEEP COPY of variable1 was done earlier, even the DEEP inner list item DID NOT CHANGE.
# For varialbe4 - because simple assignment of variable1 was done earlier, all changes to variable1 GOT REFLECTED.
```

```python
import pandas as pd
df = pd.read_csv("Sleep_Efficiency.csv")
print(df.columns)
```

#### New column creation based on some conditions
- df.map()

- In future course sessions, these functions will help with **row-wise** and **column-wise** operations on the DataFrame
    - map
    - lambda
    - apply

```python
# This is a function to categorize sleep_efficiency

def categorize_sleep_efficiency(x):
    if (x >= 0.8):
        return(1)  # Good sleep
    #else:
    elif(x < 0.8):
        return(0)  # Bad sleep
```

```python
categorize_sleep_efficiency(0.90)
```

```python
df["Sleep efficiency"].map(categorize_sleep_efficiency)
```

```python
# In this code, the data for new column is generated based on the map() function
# And that new data is assigned to a new column named "sleep_category" in the DataFrame

df["sleep_category"] = df["Sleep efficiency"].map(categorize_sleep_efficiency)
```
