### [10m] Data Preparation for Visualization

```python
# Removing the null values

df = df.dropna()

#------------------------------------------

# Correcting the column names

new_column_name_list = []

for name in df.columns:
    temp_name = name.lower()
    temp_name = temp_name.replace(" ", "_")
    new_column_name_list.append(temp_name)

df.columns = new_column_name_list

#------------------------------------------

# Data type check and fix

df["bedtime"] = pd.to_datetime(df["bedtime"], format="mixed")
df["wakeup_time"] = pd.to_datetime(df["wakeup_time"], format="mixed")

#------------------------------------------

# Sorting operation

df = df.sort_values(by="age", ascending=True)
```

### [60m] Exploratory Data Analysis

#### Matplotlib - Subplots
- Multiple plots as a single Big Box of images
- Refer: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

```python
import matplotlib.pyplot as plt

# Note: We are unpacking 2 returned variables from the subplots() function
# Refer: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

#ABCD = plt.subplots(2, 2, figsize=(10,5))  # figsize --> (width, height)
fig, axes = plt.subplots(2, 2, figsize=(10,5))  # figsize --> (width, height)

axes[0][0].scatter(df["sleep_efficiency"], df["sleep_duration"])
axes[0][1].boxplot(df["sleep_efficiency"])
axes[1][0].hist(df["age"])
axes[1][1].plot(df["age"], df["deep_sleep_percentage"])

#fig.set_figwidth(5)
#fig.set_facecolor("cyan")

plt.show()
```

#### Subplots with a Heatmap image, and another image including secondary y-axis

```python
import matplotlib.pyplot as plt
import seaborn as sns


#### Convert all Strings to Numerical type if possible, for Heatmap
df["gender"] = df["gender"].replace({"Female": 1, "Male": 2})
df["smoking_status"] = df["smoking_status"].replace({"Yes": 1, "No": 0})

df_correlations = df.corr()



# Note: We are unpacking 2 returned variables from the subplots() function
# Refer: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

fig, axes = plt.subplots(3, 2, figsize=(10,8))  # figsize --> (width, height)

axes[0][0].scatter(df["sleep_efficiency"], df["sleep_duration"])
axes[0][1].boxplot(df["sleep_efficiency"])
axes[1][0].hist(df["age"])
axes[1][1].plot(df["age"], df["deep_sleep_percentage"])

#axes[2][0]
#axes[2][1]
sns.heatmap(df_correlations, cmap="crest", ax=axes[2][0])


import copy
new_df = copy.deepcopy(df)
new_df.index = new_df["age"]
new_df["sleep_duration"].plot(ax=axes[2][1])
new_df["sleep_efficiency"].plot(secondary_y=True, ax=axes[2][1])


plt.show()
```
