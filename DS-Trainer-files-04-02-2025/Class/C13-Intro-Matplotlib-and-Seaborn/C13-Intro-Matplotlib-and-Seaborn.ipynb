### [30m] Preparing the Dataset for Visualization

```python
import pandas as pd
df = pd.read_csv("Sleep_Efficiency.csv")
```

```python
df.info()
```

#### Removing the null values

```python
df = df.dropna()
```

```python
df.info()
```

#### Correcting the column names according to the standard

```python
df.head()
```

```python
df.columns
```

```python
new_column_name_list = []

for name in df.columns:
    temp_name = name.lower()
    temp_name = temp_name.replace(" ", "_")
    new_column_name_list.append(temp_name)

print(new_column_name_list)
```

```python
df.columns = new_column_name_list
```

```python
df.columns
```

#### Data type check and fix

```python
df.info()
```

```python
df["bedtime"] = pd.to_datetime(df["bedtime"], format="mixed")
df["wakeup_time"] = pd.to_datetime(df["wakeup_time"], format="mixed")
```

```python
df.info()
```

#### Sort the DataFrame with ascending order of "age"

```python
df = df.sort_values(by="age", ascending=True)
```

### [30m] Matplotlib for Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
df.head()
```

#### Line plot

```python
# Reference: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
# Do line plot
# x axis and y axis

plt.figure(figsize=(10,4))
#plt.plot(df["age"], df["wakeup_time"])
plt.plot(df["age"], df["sleep_duration"])
plt.xlabel("Age")
plt.ylabel("Sleep Duration")
plt.title("Line Plot")
plt.savefig("line_plot.png")
```

```python
# Plotting with 2 y-axis and 1 x-axis
# primary y-axis: sleep_duration
# secondary y-axis: sleep_efficiency
# x-axis: age

# Reference: Legend for the plot - https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.legend.html

import copy
new_df = copy.deepcopy(df)

new_df.index = new_df["age"]  # Note: We have set the Age column values as index. So that the plot has the required X-axis.

new_df["sleep_duration"].plot()
plt.legend(loc='upper left')
new_df["sleep_efficiency"].plot(secondary_y=True)
#plt.legend(["Blue","Orange"], loc='upper right')
plt.legend(loc='upper right')
```

### [15m] Seaborn - Heatmap (correlations)
- Data Correlations
    - Postively correlated  (Directly proportional)
    - Negatively correlated (Inversely proportional)
    - No correlation
- **Purpose:** To understand the relationship between various columns in our DataFrame
    - Eventually, if needed, we can drop 1 or more columns before we go to the next step of Data Analysis / Machine Learning

```python
# Reference: https://seaborn.pydata.org/generated/seaborn.heatmap.html

import seaborn as sns
```

```python
# To check the DataType of our DataFrame columns
df.info()
```

```python
df.head()
```

```python
df.shape
```

#### Convert all Strings to Numerical type if possible

```python
df["gender"].value_counts()
```

```python
df["smoking_status"].value_counts()
```

```python
df["gender"] = df["gender"].replace({"Female": 1, "Male": 2})
```

```python
df["smoking_status"] = df["smoking_status"].replace({"Yes": 1, "No": 0})
```

```python
df.info()
```

```python
# towards 1 - highly positively correlated
# towards 0 - less correlation
# towards -1 - highly negatively correlated

df.corr()
```

```python
# To get the correlation among the columns of our DataFrame
df_correlations = df.corr()
```

```python
# Reference for colors: https://seaborn.pydata.org/tutorial/color_palettes.html

#sns.heatmap(df_correlations, cmap="flare")
#sns.heatmap(df_correlations, cmap="crest")
#sns.heatmap(df_correlations, cmap=["red","white","green"])

sns.heatmap(df_correlations, cmap="crest")
```
