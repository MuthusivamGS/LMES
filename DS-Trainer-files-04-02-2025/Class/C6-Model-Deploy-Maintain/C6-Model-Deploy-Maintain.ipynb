### [10m] Supervised Vs Unsupervised Machine Learning

- **Supervised Machine Learning**
    - We have a "Target feature" in our dataset
        - Eg:
            - **Goal: Good Sleep and Bad Sleep prediction**
            - Our Target feature: **"Sleep Category"**
            - Our other features: Age, Gender, Bedtime, Wakeup time, Sleep duration, REM sleep percentage, Deep sleep percentage, Light sleep percentage, Awakenings, Caffeine consumption, Alcohol consumption, Smoking status, Exercise frequency
    - We proceed to **create a ML model** with that data
    - Later we use that **ML Model to predict** the **"Sleep Category"**

----------------------------------------------------------------------------------

- **Unsupervised Machine Learning**
    - Reasons to go for that:
        - To understand the **various groups** in a dataset
            - To find the "Goal" for a Dataset in Machine Learning
    - Eg: Customer Segmentation (Client Segmentation)
    - **Demo**
        - **cluster_0**
            - Observations
                - **Sleep efficiency** is in average 11.23% greater : mean of 0.88 against 0.79 globally
                - 79.31% of the cluster has **Female for Gender** (against 50 % globally)
                - **Caffeine consumption** is in average 75.03% greater : mean of 39.66 against 22.66 globally
        - **cluster_1**
            - Observations
                - 81.94% of the cluster has **Male for Gender** (against 50 % globally)
                - **Deep sleep percentage** is in average 17.61% **greater** : mean of 62.15 against 52.84 globally
                - **Caffeine consumption** is in average 73.95% smaller : mean of 5.90 against 22.66 globally
        - **cluster_2**
            - Observations
                - **Light sleep percentage** is in average 105% greater : mean of 50.32 against 24.49 globally
                - **Deep sleep percentage** is in average 48.22% **smaller** : mean of 27.36 against 52.84 globally
                - **Sleep efficiency** is in average 25.35% smaller : mean of 0.59 against 0.79 globally
